#Merging joining and concatenating

Concatenating DataFrame using .concat() :

##Concatenating

```python
import numpy as np
import pandas as pd
```

```python
data1={'names':['noreen','asra','arifa','atiqa'],
       'age':[22,23,24,25], 'addres':['danyore','yasin','gilgit','nagar'],
       'qulification':['bs','ma','ca','fa']}

data2 = {'Name':['rida', 'noor', 'hoor', 'yasir'],
        'Age':[17, 14, 12, 52],
        'Address':['danyore', 'yasin', 'gilgit', 'nagar'],
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']}
```

```python
# Convert the dictionary into DataFrame
df = pd.DataFrame(data1,index=[0, 1, 2,3])

```

```python
df
```

```python
df1=pd.DataFrame(data2,index=[4,5,6,7])
```

```python
df1
```

```python
print(df, '\n\n' ,df1)
```

```python
# using ignore_index
res = pd.concat([df, df1], ignore_index=True)

res
```

##intersection


Concatenating DataFrame by setting logic on axes :

```python
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
```

```python
df1 = pd.DataFrame(data2, index=[2, 3, 6, 7])
```

```python
df1
```

```python
print(df, '\n\n' ,df1)
```

```python
# Now we set axes join = inner for intersection of dataframe
# applying concat with axes
# join = 'inner'
res2 = pd.concat([df, df1], axis=1, join='inner')

res2
```

##union

```python
# Now we set axes join = outer for union of dataframe.

res3=pd.concat([df, df1],axis=1,join = 'outer')
res3
```

```python
res = df.append(df1)
res
```
